## UNIXの基本

### 変数や基本コマンド (pwd, ls, cd, mkdir, rm)

```
# "%%bash" マジックコマンドでセル内でUNIXコマンドを実行できる
%%bash

# 変数の代入 (colabではこのセル内でのみ有効)
NUM=10

# "echo" コマンド : 出力する
echo "the best number is $NUM"
```

```
%%bash

# "pwd" コマンド : 現在のディレクトリのパス(位置)を表示
pwd
```

```
%%bash

# "ls" コマンド : 現在のディレクトリ下にあるファイル・ディレクトリを列挙
ls
```

```
%%bash

# "ls" コマンドの後にパスを入力することでそのパスの下にあるものを列挙
# スペースを入れる時は別の引数と見なされないように''で囲む，変数代入時は${}で囲むことがポイント
ls gdrive/'Team Drives'/coenhance_teams/team${TEAM}/content/code
```

```
%%bash

# "cd" コマンド : 指定したディレクトリに移動 (colabではセル内でのみ有効)
cd gdrive/'Team Drives'/coenhance_teams/team${TEAM}/content/code
pwd
```

```
%%bash

cd gdrive/'Team Drives'/coenhance_teams/team${TEAM}/content/code

# "mkdir" コマンド : 現在のディレクトリ下に新しいディレクトリを作る、"新しいフォルダ"と同じ
mkdir hello

ls
```

```
%%bash

cd gdrive/'Team Drives'/coenhance_teams/team${TEAM}/content/code

# "rm" コマンド : ファイルを削除、"-r" (再帰的)オプションをつけることでディレクトリごと削除が可能
rm -r hello

ls
```

## pythonの基本

### 変数の型

```
# 数値型
x = 10
y = 200
print(x+y)
```

```
print(x*y)
```

```
# 文字列型
string1 = "hello world"
string2 = 'hello world 2'
```

```
print(string1)
print(string2)
```

```
# boolean型
hello = True
```

```
print(hello)
```

```
print(not hello)
```

```
# リスト型
obj = [2, 31, 10, 0.2, 29]
```

```
print(obj)
```

```
print(obj[1])
```

```
# 辞書型 (キーと値のペアを保持)
scores = {
    "math" : 70,
    "japanese" : 20,
    "social" : 30,
    "PE" : 100
}
```

```
print(scores)
```

```
print(scores['japanese'])
```
